```python
import pandas as pd
import numpy as np
```

```python
#load data into panda

data = pd.read_csv('/content/energydata_complete.csv')
```

```python
data = data.drop(['date', 'lights'], axis = 1) 
```

```python
data.head()
```

```python
#normalizing the dataset
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

data_scaled = pd.DataFrame(scaler.fit_transform(data),  columns = data.columns)

#get features and labels
X = data_scaled.drop(columns=['Appliances'])

y = data_scaled['Appliances']
```

```python
#spliting the data into training and testing

from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print('X_train shape: {}'.format(X_train.shape))
print('y_train shape: {}'.format(y_train.shape))
print('X_test shape: {}'.format(X_test.shape))
print('y_test shape: {}'.format(y_test.shape))
```

```python
#select a sample from dataset

rdf = data_scaled[['T2', 'T6']]
```

```python
rdf.head()
```

```python
rdf.tail()
```

```python
#reshape

x= rdf['T2'].values.reshape(-1,1)
y = rdf['T6'].values.reshape(-1,1)
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
#linear model for sample dataset
from sklearn import linear_model


lregr = linear_model.LinearRegression()

# Train the model
lregr.fit(xtrain, ytrain)

# Make predictions 
pred = lregr.predict(xtest)
```

```python
#R-squared(Coefficient of determination)
from sklearn.metrics import r2_score

r2_score = r2_score(ytest, pred)
print('R-square:',(round(r2_score, 2)))
```

```python
from sklearn import linear_model

# Create linear regression object
regr = linear_model.LinearRegression()

# Train the model using the training sets
regr.fit(X_train, y_train)

# Make predictions using the testing set
energypred = regr.predict(X_test)

print("Training set score: {:.3f}".format(regr.score(X_train, y_train)))
print("Test set score: {:.3f}".format(regr.score(X_test, y_test)))
```

```python
#Mean absolute Error
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, energypred)

print('MAE:',(round(mae, 2)))

#Residual sum of squares
import numpy as np
rss = np.sum(np.square(y_test - energypred))
print('RSS:',(round(rss, 2)))

#root mean squared error
from sklearn.metrics import  mean_squared_error

rmse = np.sqrt(mean_squared_error(y_test, energypred))
print('RMSE:',(round(rmse, 3)))

#R-squared or coefficient of determination
from sklearn.metrics import r2_score

r2_score = r2_score(y_test, energypred)
print('R-squared:',(round(r2_score, 2)))
```
